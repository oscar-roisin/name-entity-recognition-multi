```python
### SWEDISH

import RDRPOS
from RDRPOS import *
import nltk

r = RDRPOSTagger()

r.constructSCRDRtreeFromRDRfile("./model/Swedish.RDR")

DICT = readDictionary("./model/Swedish.DICT")

# Tag a tokenized/word-segmented sentence
text = nltk.word_tokenize("I sin första reaktion på sovjetledarens varningar deklarerade Litauens president Vytautas Landsbergis att \"nu avvisar Gorbatjoc vår utsträckta hand\".")
# print(text)
textTest = ""

for token in text:
    textTest = textTest + " " + token

posTagged = r.tagRawSentence(DICT, textTest)
posTaggedSplit=posTagged.split()
tokens=[]
posTags=[]
print(posTaggedSplit)
for posTaggedToken in posTaggedSplit:
    splitToken = posTaggedToken.split("/")
    token = splitToken[0]
    tokens.append(token)
    pos = splitToken[1].split('.')[0]
    posTags.append(pos)
print(tokens)
print(posTags)
```
